```python
# Se encuentra que el campo ‘EDAD’ posee valores sin datos por lo cual se procede a calcular su media para imputar los valores faltantes 
df_edades= pd.DataFrame(df_hom_victimas['EDAD'])
lista_edades = list(df_edades['EDAD'])
len(lista_edades)
```

```python
# Se crea una nueva lista sin los valores 'SD' y las comas que le siguen
edades_lista = [valor for valor in lista_edades if str(valor) != 'SD' and str(valor) != ',']
len(edades_lista)
```

```python
# Se convierte los valores a números enteros y se calcula el promedio
valores_numericos = [int(valor) for valor in edades_lista]
promedio_edades = sum(valores_numericos) / len(valores_numericos) if len(valores_numericos) > 0 else 0
promedio_edades = int(promedio_edades)
print("El promedio de edades es de:", promedio_edades)
```

```python
# Se decide imputar el valor promedio de edades en donde no hay datos 
df_hom_victimas['EDAD'] = df_hom_victimas['EDAD'].replace('SD', promedio_edades)
```

```python
# Se procede a cambiar el tipo de dato del campo edad por tipo entero
df_hom_victimas['EDAD'] = df_hom_victimas['EDAD'].astype(int)
df_hom_victimas.info()
```

```python
# Se procede normalizar el nombre de todas las columnas 
nombres_normalizados = ['Id','Rol','Sexo','Edad']
df_hom_victimas.columns = nombres_normalizados
df_hom_victimas.head()
```
